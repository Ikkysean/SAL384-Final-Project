```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
import statsmodels.formula.api as smf
from pybaseball import lahman
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time
NBA = pd.read_csv("Merge.csv")
NBA = NBA[NBA["G"] >= 10]
```

```python
OLS_Formula = "DBPM ~ MP + Ht + Wt + Pos + Age + Q('DRB%') + Q('ORB%')"
OLS_BPM = smf.ols(formula = OLS_Formula, data = NBA).fit()
OLS_BPM.summary()
```

```python
Position = NBA[['Pos','DBPM']].groupby('Pos').mean()
Posbar = Position.plot.bar()
Posbar.set_title("DBPM by Position")
Posbar.set_xlabel("Position")
Posbar.set_ylabel("DBPM")
```

```python
Age = NBA[['Age','DBPM']].groupby('Age').mean()
Ageline = Age.plot.line()
Ageline.set_title("DBPM by Age")
Ageline.set_xlabel("Age")
Ageline.set_ylabel("DBPM")
```

```python
Height = NBA[['Ht','DBPM']].groupby('Ht').mean()
Heightline = Height.plot.line()
Heightline.set_title("DBPM by Height")
Heightline.set_xlabel("Height")
Heightline.set_ylabel("DBPM")
```

```python
Weight = NBA[['DRB%','DBPM']].groupby('DRB%').mean()
Weightline = Weight.plot.line()
Weightline.set_title("DBPM by Defensive Rebound")
Weightline.set_xlabel("Defensive Rebound")
Weightline.set_ylabel("DBPM")
```

```python
Weight = NBA[['ORB%','DBPM']].groupby('ORB%').mean()
Weightline = Weight.plot.line()
Weightline.set_title("DBPM by Offensive Rebound")
Weightline.set_xlabel("Offensive Rebound")
Weightline.set_ylabel("DBPM")
```

```python
Year = NBA[['Year','DBPM']].groupby('Year').mean()
Yearbar = Year.plot.bar()
Yearbar.set_title("DBPM by Season")
Yearbar.set_xlabel("Season")
Yearbar.set_ylabel("DBPM")
```
